```python
import csv
import requests
import pandas as pd
df=pd.read_csv("/home/maohuaxie/GSU/2015/df1510q.csv")
for i in range(0,len(df)):
        saveas = '-'.join([str(df.ix[i,][0]), str(df.ix[i,][1]).replace('/', '-'), str(df.ix[i,][2]).replace('/', '-'), str(df.ix[i,][3])])
        # Reorganize to rename the output filename.
        url = 'https://www.sec.gov/Archives/' + str(df.ix[i,][4]).strip()
        with open(saveas, 'wb') as f:
            f.write(requests.get('%s' % url).content)
            print(url, 'downloaded and wrote to text file')
            print(i)
```
